```python
# import the library
from geosampler import Sampler
```

```python
# initialize the sampler
place_type = "night_club"
radius = 10000 # 10km
keyword = "disco"
lang = "fr" # language is French
api_key = "<INSERT YOUR API KEY HERE>"
s = Sampler(api_key=api_key, language=lang, type_=place_type, radius=radius, keyword=keyword)
```

```python
# retrieve population
pop = s.nearby_search(locations=['44.8361,-0.5808'], extra_details=True)
```

```python
# some records are from localities close by to Bordeux
pop.head(3)
```

```python
# filter records with locality is Bodeaux and exclude those in the vicinity e.g. Le Haillan
pop = pop[pop['locality']=='Bordeaux']
# population size
pop.shape[0]
```

```python
# this is how to create an stratefied sample
ss = s.stratified_sample(population=pop, n=3, columns=['postal_code'])
ss.shape
```

```python
# it is also possible to use multiple map tiles as layers
m = s.map(location=(44.836, -0.5808), map_tiles=["Google Maps", "Google Terrain", "Google Satellite", "Google Satellite Hybrid"], datasets=[pop, ss], colors=['purple', 'blue'], icons=['fa-music']*2, zoom_start=13)
m
```

```python
# for reproducibility, both the population and the stratified sample are saved
pop['place_id'].to_csv('night_club_bordeaux_france_pop.csv', index=False)
ss['place_id'].to_csv('night_club_bordeaux_france_ss.csv', index=False)
```

```python
# save the map as htlm
m.save('night_club_bordeaux_france.html')
```
