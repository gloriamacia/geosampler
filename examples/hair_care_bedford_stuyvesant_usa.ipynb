```python
# import library
from geosampler import *
```

```python
# initialize the sampler
# in this example we are not using the keyword but we will filter later by name
place_type = "hair_care"
radius = 1000 # 1km
lang = "en"
api_key = "<INSERT YOUR API KEY HERE>"
s = Sampler(api_key=api_key, language=lang, type_=place_type, radius=radius)
```

```python
pop = s.nearby_search(locations=['40.689133, -73.941570'], extra_details=True)
```

```python
pop.shape[0]
```

```python
pop.head()
```

```python
# there are some barbershops but also hair salons we need to get rid off
# we exclude any places that do not contain the string "barber" in their names
pop = pop[pop['name'].apply(lambda x : "barber" in x.lower())]
pop.shape[0]
```

```python
m = s.map(location=(40.688552, -73.939596), datasets=[pop], colors=['red'], icons=['fa-scissors'], zoom_start=14)
m
```

```python
# save population as CSV
pop['place_id'].to_csv('hair_care_bedford_stuyvesant_usa_pop.csv', index=False)
```

```python
# save map as html
m.save('map_hair_care_bedford_stuyvesant_usa.html')
```
