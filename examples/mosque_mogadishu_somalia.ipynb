```python
# import library
from geosampler import Sampler
```

```python
place_type = "mosque"
radius = 5000 # 5km
lang = "en" # you could also try Arabic ("ar"), full list supported languages: https://developers.google.com/maps/faq#languagesupport
api_key = "<INSERT YOUR API KEY HERE>"
s = Sampler(api_key=api_key, language=lang, type_=place_type, radius=radius)
```

```python
# Mogadishu is located at 2°4' North, 45°22' East (2.06667, 45.36667).
# it is a large city so we are sampling a few points from the city area, as may have >60 mosques
# this may take a couple minutes as we are explorig a large area
# the more places you retrieve the higher the cost, consider this before running this cell. 
pop = s.nearby_search(locations=['2.066177, 45.370379', '2.084649, 45.348914', '2.062304, 45.309562', '2.038172, 45.318506', '2.019742, 45.272329', '2.036211, 45.250700', '2.054955, 45.342241', '2.042560, 45.345589', '2.038486, 45.293688', '2.026171, 45.309630'], extra_details=True)
```

```python
# filter records with locality is Mogadishu and exclude those in the vicinity 
pop = pop[pop['locality']=='Mogadishu']
# total population size
pop.shape[0]
```

```python
pop.head(3)
```

```python
# we randomly sample 50 mosques
rs = s.random_sample(population=pop, n=50)
print(rs.shape[0])
```

```python
m = s.map(location=(2.06667, 45.36667), datasets=[pop, rs], colors=['green', 'darkgreen'], icons=['fa-moon-o']*2, zoom_start=13)
m
```

```python
pop['place_id'].to_csv('mosque_mogadishu_somalia_pop.csv', index=False)
rs['place_id'].to_csv('mosque_mogadishu_somalia_rs.csv', index=False)
```

```python
m.save('mosque_mogadishu_somalia.html')
```
